```python
import numpy as np
import matplotlib.pyplot as plt

# Set the detection rates
detection_rates = {
    0: 0.9595637137536579,
    1: 0.6934673366834171,
    2: 0.718663697104677,
    3: 0.4724128710223728,
    4: 0.7026644241834116,
    5: 0.8849114014676929,
    6: 0.9793209319125881,
    7: 0.9029562404528709,
    8: 0.9978401727861771,
    9: 0.9995515695067264,
}

# Create a list of attack classes
attack_classes = list(detection_rates.keys())

# Convert the detection rates to a NumPy array
detection_rates_array = np.array(list(detection_rates.values()))

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(attack_classes, detection_rates_array, color='lightpink', edgecolor='black')

# Add value labels on top of each bar
for bar in plt.gca().patches:
    plt.annotate('{:.4f}%'.format(bar.get_height() * 100),
                (bar.get_x() + bar.get_width() / 2,
                 bar.get_height()),
                ha='center',
                va='center',
                fontsize=10)

# Set the x-axis label
plt.xlabel('Attack Class')

# Set the y-axis label
plt.ylabel('Detection Rate')

# Set the title
plt.title('Detection Rates of Hybrid RF-LSTM Model')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the detection rates
detection_rates = {
    0: 95.95637137536579,
    1: 69.34673366834171,
    2: 71.8663697104677,
    3: 47.24128710223728,
    4: 70.26644241834116,
    5: 88.49114014676929,
    6: 97.93209319125881,
    7: 90.29562404528709,
    8: 99.78401727861771,
    9: 99.95515695067264,
}

# Create a list of attack classes
attack_classes = list(detection_rates.keys())

# Convert the detection rates dictionary to a NumPy array
detection_rates_array = np.array(list(detection_rates.values()))

# Create a list of tuples, where each tuple contains one element from each array
data = list(zip(detection_rates_array, attack_classes))

# Create a violin plot
fig, ax = plt.subplots(figsize=(8, 6))
violin_parts = ax.violinplot(data)

# Set the x-axis label
ax.set_xlabel('Class')

# Set the y-axis label
ax.set_ylabel('Detection Rate (%)')

# Set the title
ax.set_title('Detection Rates of Hybrid RF-LSTM Model')

# Set the color of the violin plot
for pc in violin_parts['bodies']:
    pc.set_facecolor('#ff9999')  # Change the color to pink
    pc.set_edgecolor("black")
    pc.set_alpha(0.75)

# Show the bar values
for i, (rate, _) in enumerate(data):
    ax.text(i + 1, rate, f'{rate:.4f}', ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the detection rates
detection_rates = {
    0: 99.70625138386596,
    1: 99.98385771307818,
    2: 99.97656776310009,
    3: 99.90001323354262,
    4: 99.99852694222667
    }


# Create a list of attack classes
attack_classes = list(detection_rates.keys())

# Convert the detection rates dictionary to a NumPy array
detection_rates_array = np.array(list(detection_rates.values()))

# Create a list of tuples, where each tuple contains one element from each array
data = list(zip(detection_rates_array, attack_classes))

# Create a violin plot
fig, ax = plt.subplots(figsize=(8, 6))
violin_parts = ax.violinplot(data)

# Set the x-axis label
ax.set_xlabel('Class')

# Set the y-axis label
ax.set_ylabel('Detection Rate (%)')

# Set the title
ax.set_title('Detection Rates of Hybrid RF-LSTM Model')

# Set the color of the violin plot
for pc in violin_parts['bodies']:
    pc.set_facecolor('#ff9999')  # Change the color to pink
    pc.set_edgecolor("red")
    pc.set_alpha(0.75)

# Show the bar values
for i, (rate, _) in enumerate(data):
    ax.text(i + 1, rate, f'{rate:.4f}', ha='center', va='bottom')

# Show the plot
plt.show()
```
