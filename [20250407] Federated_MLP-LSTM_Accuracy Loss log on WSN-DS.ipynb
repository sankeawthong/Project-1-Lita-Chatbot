```python
def evaluate(self, parameters, config):
    self.model.set_weights(parameters)
    loss, acc = self.model.evaluate(self.X_test, self.y_test, verbose=0)

    # Log accuracy per round
    round_accuracies.append(acc)

    print(f"Client {self.client_id} | Round {config['server_round']} | Accuracy: {acc:.8f}")
    return loss, len(self.X_test), {"accuracy": acc}
```

```python
import matplotlib.pyplot as plt

# Initialize round_accuracies before using it
round_accuracies = []  # This line is crucial

# Only plot if values are logged
if round_accuracies:
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, len(round_accuracies) + 1), round_accuracies, marker='o', linewidth=2)
    plt.title("ğŸ“ˆ Federated Learning Accuracy per Round")
    plt.xlabel("Federated Round")
    plt.ylabel("Client Accuracy")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
