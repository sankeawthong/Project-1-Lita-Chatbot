```python
import pandas as pd

df_tr = pd.read_csv("/content/CIC_IoMT_2024_WiFi_MQTT_train.csv", usecols=["label"])
df_ho = pd.read_csv("/content/CIC_IoMT_2024_WiFi_MQTT_test.csv", usecols=["label"])

print("Train benign count:",
      df_tr["label"].astype(str).str.contains("benign", case=False, na=False).sum(),
      " / ", len(df_tr))

print("Holdout benign count:",
      df_ho["label"].astype(str).str.contains("benign", case=False, na=False).sum(),
      " / ", len(df_ho))
```

```python
import os

output_dir = "/content/paper_exports/strict_run"

if os.path.exists(output_dir):
    for root, dirs, files in os.walk(output_dir):
        level = root.replace(output_dir, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print(f'{indent}{os.path.basename(root)}/')
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print(f'{subindent}{f}')
else:
    print(f"Directory '{output_dir}' does not exist.")
```
