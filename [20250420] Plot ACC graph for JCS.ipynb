```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Data
models = ['RF', 'LSTM', 'Hybrid RF-LSTM', 'Hybrid LSTM-RF']
datasets = ['WSN-DS', 'WSN-BFSF', 'UNSW-NB15']
accuracy = {
    'WSN-DS': [97.77, 98.21, 99.92, 98.81],
    'WSN-BFSF': [96.06, 98.53, 99.98, 99.53],
    'UNSW-NB15': [72.59, 73.18, 86.89, 77.37]
}

# Plot configuration
bar_width = 0.2
x = np.arange(len(models))
colors = ['#4C72B0', '#55A868', '#C44E52']  # Professional color palette

fig, ax = plt.subplots(figsize=(10, 6))

# Plot each dataset
for i, (dataset, values) in enumerate(accuracy.items()):
    bars = ax.bar(x + i * bar_width, values, width=bar_width, label=dataset, color=colors[i])
    # Add value labels
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval:.2f}%', ha='center', va='bottom', fontsize=8)

# Axes labels and ticks
ax.set_xlabel('Model Configuration', fontsize=12)
ax.set_ylabel('Accuracy (%)', fontsize=12)
ax.set_title('Comparative Accuracy across Models and Datasets', fontsize=14)
ax.set_xticks(x + bar_width)
ax.set_xticklabels(models, fontsize=10)
ax.set_ylim(65, 102)
ax.legend(title='Dataset')

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save the figure
output_path = "/mnt/data/grouped_bar_accuracy_labeled.png"

# Create the directory if it doesn't exist
os.makedirs(os.path.dirname(output_path), exist_ok=True)

plt.savefig(output_path, dpi=400)
plt.show()

output_path
```
